```python
from abc import ABC, abstractmethod

class Vehicle (ABC) :

    @abstractmethod
    def calculate_fare (self, distance) :
        pass

    @abstractmethod
    def get_vehicle_type (self):
        pass

class UberX (Vehicle) :

    def calculate_fare (self, distance) :
        return distance * 10

    def get_vehicle_type (self) :
        return "UberX"

uberx = UberX ()
print (uberx . calculate_fare (5) )
```

Why can't we instantiate Vehicle?
- Because it is an abstract class with abstract methods, so Python does not allow creating objects from it.

What happens if UberX does not implement get_vehicle_type()?
- Python raises an error because the subclass did not implement all required abstract methods.

Is ABC enforcing structure at runtime or compile-time?
- It enforces structure at runtime because Python checks abstract methods when creating objects.
Why is this useful for large

companies like Uber?
- It ensures all vehicle types follow the same contract and prevents missing required methods.

```python
class UberBike (Vehicle) :

    def calculate_fare (self, distance) :
        return distance * 5

    def get_vehicle_type (self) :
        return "UberBike"

uberx = UberX ()
bike = UberBike ()

print (isinstance (uberx, Vehicle) )
print (isinstance (bike, Vehicle) )
```

Is UberBike automatically a Vehicle?
- Yes

What happens if we remove (Vehicle) from UberBike?
- UberBike stops being a Vehicle and no longer follows the required contract.

Is inheritance modeling IS-A or HAS-A?
- Inheritance models an IS-A relationship.

```python
vehicles = [UberX (), UberBike () ]
for vehicle in vehicles :
    print (vehicle . get_vehicle_type (),
           vehicle. calculate_fare (10))
```

Where does runtime polymorphism occur?
- When the loop calls the same method on different vehicle objects and Python selects the correct version at runtime.

What would happen if method names differed?
- Polymorphism would break because the shared method call would not exist in all classes.

Why is this better than using if-else checks?
- It makes the system easier to extend and avoids modifying code when adding new vehicle types.


```python
class ElectricMixin :

    def charge_battery (self) :
        return " Charging battery ... "

class UberGreen (ElectricMixin, UberX) :
    pass

green = UberGreen ()
print (green . get_vehicle_type ())
print (green . charge_battery () )
```

Why is ElectricMixin listed first?
- Because Python's method resolution order gives priority to the first parent.

Is ElectricMixin modeling IS-A?
- No,

What happens if both parents define the same method?
- Python uses the method from the first parent in the inheritance list.


```python
class Driver :

    def __init__ (self, name, vehicle) :
        self. name = name
        self. vehicle = vehicle

    def start_trip (self, distance) :
        fare = self . vehicle. calculate_fare (distance)
        return f" {self . name} driving {self . vehicle . get_vehicle_type () } - Fare : {fare }"

driver1 = Driver ( "Alice", UberX () )
driver2 = Driver ( "Bob", UberBike () )

print (driver1. start_trip (10))
print (driver2. start_trip (10))
```

Is Driver inheriting from Vehicle?
- No

What relationship is this modeling?
- A HAS-A relationship.

Why is composition flexible?
- Because you can change the vehicle without modifying the Driver class.

```python
class UberApp :
    def request_ride (self, driver, distance) :
        return driver. start_trip (distance)

app = UberApp ()
print (app.request_ride (driver1, 5))
```

How is dependency different from composition?
- Dependency is temporary use, while composition is a permanent part-whole relationship.

Does UberApp permanently store Driver?
- No

Why is loose coupling important in real systems?
- It allows components to change independently and keeps the system maintainable.
